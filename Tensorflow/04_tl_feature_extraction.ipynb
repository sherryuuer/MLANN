### 迁移学习
* 什么是迁移学习

迁移学习（Transfer Learning）是一种机器学习方法，它利用在一个任务上学到的知识来改善在一个相关但不同的任务上的性能。在传统的机器学习中，通常假设训练数据和测试数据是独立同分布的，而迁移学习则研究如何在源领域（source domain）上学到的知识，如何应用于目标领域（target domain），尤其是当源领域和目标领域的分布不完全相同时。

迁移学习的基本思想是，通过在一个相关任务上学习到的知识，可以加速和改善在新任务上的学习过程。迁移学习的主要优势在于，当目标任务的数据量相对较小或者标注成本较高时，可以通过借助源任务上的大量数据和知识来提高模型的泛化性能。

有两种主要的迁移学习方法：

1. **特征提取迁移学习：** 在这种方法中，使用在源任务上学到的模型的底层或中间层的特征提取能力。这些特征通常能够捕捉到通用的数据表示，对于目标任务也是有用的。例如，使用在大规模图像分类任务上预训练的卷积神经网络（CNN）的特征提取能力，然后在目标任务上训练一个简单的分类器。

简单来说就是，只改变output层的内容，让模型适合你的问题。

2. **模型微调（Fine-tuning）：** 在这种方法中，使用在源任务上训练好的模型的权重作为初始权重，然后在目标任务上进行进一步的训练。模型微调可以在整个模型上进行，也可以只微调部分层次。例如，通过微调在语言模型上预训练的循环神经网络（RNN）来执行一个文本分类任务。

迁移学习在计算机视觉、自然语言处理等领域取得了显著的成功，特别是在深度学习的框架下，预训练的大型模型如BERT、GPT等被广泛应用于各种下游任务。

越靠近输入层，越是能学到最显著的特征，就像我们看图一样，一开始看到猫狗，要辨别它是什么的时候最先关注的是他们最显著的特征，越靠近输出层，越是那些微妙的差别，比如牙齿的形状之类的细小不同。

* 什么是resnet
ResNet（Residual Network）是由微软研究院提出的一种深度神经网络架构，旨在解决深度神经网络训练过程中的梯度消失和梯度爆炸等问题。这一问题在训练非常深的神经网络时尤为显著，因为深度增加会导致梯度难以传播到网络的较早层，从而使得网络难以收敛。

ResNet通过引入残差块（Residual Block）来解决这个问题。在传统的神经网络中，每个层的输出都是通过非线性激活函数（如ReLU）作用于前一层的输出得到的。而在ResNet的残差块中，输出不仅与前一层的输出相关，还与前一层的输入相关，通过添加残差连接实现。

具体而言，一个残差块的计算公式为：

输出 = 激活(权重 * 输入 + 输入）

这里的ast表示卷积操作，激活函数通常是ReLU。这种残差块的设计允许网络学习对于输入的残差，而不是直接学习映射。残差块的引入使得网络可以更轻松地学习恒等映射，避免了梯度消失和梯度爆炸问题。

ResNet的架构使得可以训练非常深的网络，例如ResNet可以有数百层，而仍然能够有效地进行训练。ResNet在深度学习领域取得了显著的成功，并在图像分类、目标检测等任务上取得了 state-of-the-art 的性能。

* 什么是模型微调
模型微调（Fine-tuning）是深度学习中一种常见的训练策略，它指的是在一个已经在大规模数据集上预训练好的模型基础上，对模型进行进一步的训练以适应特定任务或数据集。微调通常发生在两个阶段：

1. **预训练阶段（Pretraining）：** 在这个阶段，模型首先在一个大规模的通用数据集（如ImageNet）上进行训练。这个过程使模型学到了通用的特征表示和模式，使其成为一个强大的特征提取器。

2. **微调阶段（Fine-tuning）：** 接下来，使用在预训练阶段学到的权重作为初始权重，将模型在目标任务的特定数据集上进行额外的训练。在微调阶段，模型的参数会根据目标任务的数据进行调整，以适应任务的特定要求。

微调的优势包括：

- **迁移学习：** 通过在通用数据集上进行预训练，模型可以学到通用的特征表示，这有助于在新任务上获得更好的性能。这种思想是迁移学习的基础。

- **节省计算资源：** 预训练可以在大规模数据上进行，利用更多的标注数据和计算资源。而微调则在目标任务上进行，避免了从零开始训练模型的需要。

在微调过程中，通常需要调整学习率、解冻一些层以允许权重更新，或者根据目标任务的特定要求进行一些其他调整。微调的性能通常取决于预训练模型的质量、目标任务的相似性以及微调阶段的参数设置。

* 什么是EfficientNet以及它后面的号码是什么
EfficientNet是一系列由Google提出的卷积神经网络架构，其后面的数字（如EfficientNetB0、EfficientNetB1等）表示不同的模型大小或复杂度。这些数字主要用于指示网络的规模，通常是通过扩展网络的宽度、深度和分辨率来实现的。

具体而言，EfficientNet使用了一个复合系数（compound coefficient），用于确定每个维度的缩放因子。这个复合系数被表示为一个元组（phi，alpha，beta，gamma），其中：

- **phi：** 控制网络的整体深度（层数）。较大的phi值通常对应于更深的网络。

- **alpha：** 控制网络的宽度（通道数）。较大的alpha值对应于更宽的网络。

- **beta：** 控制输入图像的分辨率。较大的beta值对应于更高分辨率的输入图像。

- **gamma：** 是一个扩展系数，用于控制网络的复杂度。通常情况下，gamma的值为2。

这些系数被用于扩展基础网络结构，以获得不同复杂度和性能的EfficientNet模型。例如，EfficientNetB0是基础模型，而EfficientNetB1、EfficientNetB2等则是通过增加深度和宽度来扩展的版本。这样的设计使得EfficientNet系列可以在计算资源有限的情况下，更灵活地选择适合任务需求的模型规模。

```python
# create data loaders(preparing the data)
# setup data input
from tensorflow.keras.preprocessing.image import ImageDataGenerator

IMAGE_SHAPE = (224, 224)
BATCH_SIZE = 32

train_dir = "10_food_classes_10_percent/train/"
test_dir = "10_food_classes_10_percent/test/"

train_datagen = ImageDataGenerator(rescale=1/255.)
test_datagen = ImageDataGenerator(rescale=1/255.)

print("Training images:")
train_data_10_percent = train_datagen.flow_from_directory(train_dir,
                                                         target_size=IMAGE_SHAPE,
                                                         batch_size=BATCH_SIZE,
                                                         class_mode="categorical")
print("Testing images:")
test_data = test_datagen.flow_from_directory(test_dir,
                                            target_size=IMAGE_SHAPE,
                                            batch_size=BATCH_SIZE,
                                            class_mode="categorical")
```

### callback
在机器学习中，"callback" 是指在模型训练的过程中，根据模型在训练集和验证集上的表现执行一些额外的操作或者函数。这可以帮助监控训练的进展，进行模型的动态调整，以及在满足特定条件时执行一些操作。Callbacks 在许多机器学习框架中都有广泛的应用，其中最常见的是在深度学习框架中，如TensorFlow和PyTorch。

以下是一些典型的 callback 应用：

1. **Early Stopping（早停）：** Early Stopping 是一种常见的 callback，它会在模型在验证集上的性能不再提升时停止训练，从而防止模型过拟合。

2. **Model Checkpoint（模型保存）：** 在训练过程中，定期保存模型的参数。这对于长时间的训练任务很有用，以防止在训练中断时丢失模型的进展。

3. **Learning Rate Scheduler（学习率调度器）：** 根据训练的进度动态调整学习率。这可以帮助加速模型收敛或提高模型在验证集上的性能。

4. **TensorBoard（可视化）：** 使用TensorBoard callback可以将训练过程中的指标和其他信息可视化，以便更好地理解模型的行为。

5. **Custom Callbacks（自定义回调）：** 用户可以根据特定需求创建自定义的回调函数。例如，执行一些特定任务，或者在训练过程中记录额外的信息。

在实际应用中，这些 callback 可以结合使用，根据任务的需要进行组合。使用 callback 可以使模型训练的过程更加灵活，同时也提高了模型训练的效率和效果。

```python
# create tensorboard callback (functionized because need to create a new one for each model)
import datetime
def create_tensorboard_callback(dir_name, experiment_name):
    log_dir = dir_name + "/" + experiment_name + "/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
    tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=log_dir)
    print(f"Saving Tensorboard log files to :{log_dir}")
    return tensorboard_callback
```

```python
import tensorflow as tf
import tensorflow_hub as hub
from tensorflow.keras import layers
```

```python
# Resnet 50 V2 feature vector
resnet_url = "https://tfhub.dev/google/imagenet/resnet_v2_50/feature_vector/4"

# Original: EfficientNetB0 feature vector (version 1)
efficientnet_url = "https://tfhub.dev/tensorflow/efficientnet/b0/feature-vector/1"

# # New: EfficientNetB0 feature vector (version 2)
# efficientnet_url = "https://tfhub.dev/google/imagenet/efficientnet_v2_imagenet1k_b0/feature_vector/2"
```

```python
# create a model function
def create_model(model_url, num_classes=10):
    """
    Take a tensorflow mdoel url, and create a keras model layer with it.

    Args:
    model_url (str): A tensorflow feature extrection url.
    num_classes: Num of output neurons in output layer, should be equal to number of target classes, default 10.

    Returns:
    an uncompiled keras Sequential model with model_url as feature
    extractor layer and Dense output layer with num_classes outputs.
    """
    feature_extractor_layer = hub.KerasLayer(model_url,
                                            trainable=False, # freeze the underlying patterns
                                            name="feature_extraction_layer",
                                            input_shape=IMAGE_SHAPE + (3,))
    model = tf.keras.Sequential([
        feature_extractor_layer,
        layers.Dense(num_classes, activation="softmax", name="output_layer")
    ])

    return model
```

```python
# create and compile model
resnet_model = create_model(resnet_url, num_classes=train_data_10_percent.num_classes)
# compile
resnet_model.compile(loss="categorical_crossentropy",
                     optimizer=tf.keras.optimizers.legacy.Adam(),
                     metrics=["accuracy"])
```

```python
# fit the model
resnet_history = resnet_model.fit(train_data_10_percent,
                                  epochs=5,
                                  steps_per_epoch=len(train_data_10_percent),
                                  validation_data=test_data,
                                  validation_steps=len(test_data),
                                  callbacks=[create_tensorboard_callback(dir_name="model_logs",
                                                                         experiment_name="resnet_model_tl")])
```

更少的数据得到更好的模型，省资源，结果好。这就像是一个有经验的老手，遇到小儿科问题，只需要看两眼书就懂了是一个道理。

```python
# check loss curve
import matplotlib.pyplot as plt
def plot_loss_curves(history):
    """
    Returns separate loss curves for training and validation metrics.
    """
    loss = history.history["loss"]
    val_loss = history.history["val_loss"]

    accuracy = history.history["accuracy"]
    val_accuracy = history.history["val_accuracy"]

    epochs = range(len(history.history["loss"]))

    # plot loss
    plt.plot(epochs, loss, label="training_loss")
    plt.plot(epochs, val_loss, label="val_loss")
    plt.title("Loss")
    plt.xlabel("Epochs")
    plt.legend()

    # plot accuracy
    plt.figure()
    plt.plot(epochs, accuracy, label="training_accuracy")
    plt.plot(epochs, val_accuracy, label="val_accuracy")
    plt.title("Accuracy")
    plt.xlabel("Epochs")
    plt.legend();
```

```python
plot_loss_curves(resnet_history)
```

使用了更少的参数，但是更好的结果。高效学习？？！！

EfficientNet 相对于 ResNet 在结果上更好的原因可以归结为其更加有效的架构设计，具体而言有以下几个方面：

1. **Compound Scaling（复合缩放）：** EfficientNet 使用了复合缩放的方法，通过同时增加网络的深度、宽度和分辨率，以实现更好的性能。这种缩放方法使得网络在不同维度上都能够更好地适应数据，相对于简单地增加网络深度而言，更加全面。

2. **MBConv（Mobile Inverted Bottleneck Convolution）结构：** EfficientNet 使用了一种称为 MBConv 的模块，它结合了轻量级网络设计和深度残差连接的优势。这种结构使得网络能够在保持有效性的同时，更加灵活地学习不同任务的特征。

3. **AutoML 搜索空间：** EfficientNet 的架构是通过 AutoML（自动机器学习）搜索空间中的搜索算法来得到的，这意味着网络结构是通过自动搜索而来的，而不是手工设计。这有助于找到更优的网络结构，适应于不同的任务。

4. **参数效率：** 相较于一些更深、更复杂的网络，EfficientNet 在参数使用上更加高效。这意味着在相同的计算资源下，EfficientNet能够获得更好的性能。

5. **迁移学习效果：** 由于 EfficientNet 在大规模的数据集上进行了预训练，它能够更好地适应各种视觉任务，特别是当迁移到新任务时，其性能可能更加优越。

需要注意的是，网络性能的优劣还取决于具体的任务、数据集以及资源的使用情况。在某些情况下，ResNet 仍然可能是一个合适的选择，特别是在计算资源有限的情况下。
